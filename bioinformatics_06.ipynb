## step1 å­¦ç¿’ã«ä½¿ã†ãƒ‡ãƒ¼ã‚¿ã®æº–å‚™
mnistã‹ã‚‰ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã™ã‚‹


*   keras.load_dataã§mnistã®ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã™ã‚‹
*   ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã—ãŸãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã¯ ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ç”¨ã¨ãƒãƒªãƒ‡ãƒ¼ã‚·ãƒ§ãƒ³ç”¨ã«åˆ†ã‹ã‚Œã¦ã„ã‚‹
*   ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã¯ ç”»åƒãƒ‡ãƒ¼ã‚¿(x)ã¨ æ­£è§£ãƒ©ãƒ™ãƒ«(y)ã‚’æŒã¤ã€‚æ­£è§£ãƒ©ãƒ™ãƒ«ã¯ç”»åƒãŒç¤ºã™æ•°å­—ã§ã‚ã‚‹ã€‚

```python
import keras
from keras.datasets import mnist

# mnist ã‹ã‚‰ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã¨ãƒãƒªãƒ‡ãƒ¼ã‚·ãƒ§ãƒ³ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã™ã‚‹
# xã¯ç”»åƒãƒ‡ãƒ¼ã‚¿ã€yã¯ç”»åƒã®åˆ¤å®šãƒ©ãƒ™ãƒ«
# x_trainãƒ‡ãƒ¼ã‚¿ã¯28*28ãƒ”ã‚¯ã‚»ãƒ«ç”»åƒãŒ60000æš (60000, 28, 28)
# x_testãƒ‡ãƒ¼ã‚¿ã¯28*28ãƒ”ã‚¯ã‚»ãƒ«ç”»åƒãŒ10000æš (10000, 28, 28)
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### load_dataã®ä¸­èº«

```python
x_train[0]
```

ã“ã®ç”»åƒã®æ­£è§£ãƒ©ãƒ™ãƒ« y_train[0] ã¯ 5

### ãƒ‡ãƒ¼ã‚¿ã®å‰å‡¦ç†
æ­£è§£ãƒ©ãƒ™ãƒ«(y)ã‚’ 10å€‹ã®ç¢ºç‡ãƒ‡ãƒ¼ã‚¿ ã®å½¢å¼ã«å¤‰æ›ã™ã‚‹ã€‚ä»Šã®æ­£è§£ãƒ©ãƒ™ãƒ«ã¯ 1ã€œ9ã®ãƒ©ãƒ™ãƒ«ã«ãªã£ã¦ãŠã‚Šã€å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã®å‡ºåŠ›å½¢å¼ã§ã‚ã‚‹10å€‹ã®ç¢ºç‡ãƒ‡ãƒ¼ã‚¿ã«åˆã‚ãªã„ãŸã‚ã€‚
ä¾‹: [0, 1, ..., 0]ã€‚å·¦ã‹ã‚‰æ•°å­—1,2, ... 9ã®ç¢ºç‡
ç”»åƒãƒ‡ãƒ¼ã‚¿(x)ã¯æ‰±ã„ã‚„ã™ã„ã‚ˆã†ã« 0~1ã®é–“ã®æ•°å€¤ã«ã™ã‚‹ã€‚ä»Šã¯ min:0(ç™½) ã€œ max:255(é»’)ã‚’è¡¨ã™ã€‚

```python
from keras.utils import to_categorical

# mnistã®ç”»åƒãƒ‡ãƒ¼ã‚¿ã¯ã‚°ãƒ¬ãƒ¼ã‚¹ã‚±ãƒ¼ãƒ«ç”»åƒã§ã‚ã‚Šã€min:0(ç™½)~max:255(é»’)ã‚’è¡¨ã™ã€‚å…¥åŠ›ã‚’0~1ã®ç¯„å›²ã«ã—ãŸæ–¹ãŒéƒ½åˆãŒã‚ˆã„ã®ã§255ã§å‰²ã‚‹ã€‚
X_train = x_train.reshape(60000, 28,28,1)/255
X_test = x_test.reshape(10000, 28,28,1)/255
# å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã®å‡ºåŠ›ãŒ 10å€‹ã®ç¢ºç‡ãƒ‡ãƒ¼ã‚¿(ä¾‹: [0, 1, ..., 0])ã«ãªã‚‹ã®ã§ã€ä»Šã®å½¢å¼(ç”»åƒãŒ0~9ã®ã†ã¡ã®ã©ã®æ•°å­—ã‹)ã¯ã‹ã¿åˆã‚ãªã„ã€‚ãã“ã§å½¢å¼ã‚’ãã‚ãˆã‚‹ãŸã‚ã«å¤‰æ›ã™ã‚‹
Y_train = to_categorical(y_train, 10)
Y_test = to_categorical(y_test, 10)
```

## step2 å­¦ç¿’
### å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã‚’ç”¨æ„ã™ã‚‹

kerasã®Sequentialãƒ¢ãƒ‡ãƒ«ã‚’ä½¿ã£ã¦å­¦ç¿’ãƒ¢ãƒ‡ãƒ«ã‚’ä½œæˆã™ã‚‹
*   Sequentialãƒ¢ãƒ‡ãƒ«ã¯ãƒ‹ãƒ¥ãƒ¼ãƒ©ãƒ«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ã‚’æ§‹ç¯‰ã™ã‚‹ãŸã‚ã®1ã¤ã®æ–¹æ³•
*   ãƒ‹ãƒ¥ãƒ¼ãƒ©ãƒ«ãƒãƒƒãƒˆãƒ¯ãƒ¼ã‚¯ã¯è¤‡æ•°ã®å±¤ï¼ˆãƒ¬ã‚¤ãƒ¤ãƒ¼ï¼‰ã§æ§‹æˆã•ã‚Œã‚‹ã€‚Sequentialã§ã¯å±¤ã‚’1ã¤ãšã¤ç©ã¿é‡ã­ã¦æ§‹ç¯‰ã—ã¦ã„ãã€‚


```python
from keras.models import Sequential
from keras.layers import Dense , Activation, Convolution2D , MaxPooling2D, Flatten

model = Sequential()
model.add(Convolution2D(10, (5,5), input_shape=(28,28,1), activation="relu")) # 2Dç•³ã¿è¾¼ã¿å±¤
model.add(MaxPooling2D(2,2)) # Maxãƒ—ãƒ¼ãƒªãƒ³ã‚°å±¤
model.add(Convolution2D(10, (5,5), padding="same"))
model.add(MaxPooling2D(2,2))
model.add(Convolution2D(10, (5,5), padding="same"))
model.add(MaxPooling2D(2,2))
model.add(Flatten()) # å¹³å¦åŒ–å±¤ã€‚å…¥åŠ›ç”»åƒã‚’ï¼‘æ¬¡å…ƒé…åˆ—ã«å¤‰æ›ã—ã¦å‡ºåŠ›ã™ã‚‹ã€‚å…¨çµåˆå±¤ã«æ¥ç¶šã™ã‚‹å‰ã«ä½¿ã‚ã‚Œã‚‹ã€‚
model.add(Dense(10, activation="softmax")) # å…¨çµåˆå±¤
```

### ç•³ã¿è¾¼ã¿å±¤(Convolution)ã¨ã¯
*   ç”»åƒã®ç‰¹å¾´ç‚¹ã‚’è¦‹ã¤ã‘ã‚‹ã“ã¨
*   ç”»åƒã«ã‚«ãƒ¼ãƒãƒ«ã‚’å……ã¦ã‚‹ã“ã¨ã§ç‰¹å¾´ç‚¹(ç‰¹å¾´ãƒãƒƒãƒ—)ã‚’å¾—ã‚‹
*   ã‚«ãƒ¼ãƒãƒ«ã®æ•°å­—ã«ã‚ˆã£ã¦æŠ½å‡ºã§ãã‚‹ç‰¹å¾´ãŒå¤‰ã‚ã‚‹ã€‚
*   CNNã«ãŠã„ã¦ã‚«ãƒ¼ãƒãƒ«ãŒé‡ã¿ã§ã‚ã‚Šã€ã‚ˆã„CNNãƒ¢ãƒ‡ãƒ«ã¨ã¯æœ€é©ãªã‚«ãƒ¼ãƒãƒ«ã‚’æŒã£ãŸãƒ¢ãƒ‡ãƒ«ã®ã“ã¨ã‚’è¨€ã†ã€‚



### ãƒ—ãƒ¼ãƒªãƒ³ã‚°å±¤(Pooling)ã¨ã¯
*   è­˜åˆ¥ã«ä¸å¿…è¦ãªç‰¹å¾´é‡ã‚’ããè½ã¨ã™
*   è­˜åˆ¥ã«å¿…è¦ãªç‰¹å¾´é‡ã‚’æŠ½å‡ºã™ã‚‹å‡¦ç†

### å…¨çµåˆå±¤(Dense)ã¨ã¯


*   ç•³ã¿è¾¼ã¿å±¤ã‚‚ã€ãƒ—ãƒ¼ãƒªãƒ³ã‚°å±¤ã‚‚ã€å‡ºåŠ›ã¯ç”»åƒã®ã¾ã¾(ç‰¹å¾´ãƒãƒƒãƒ—)ã§ã‚ã‚‹ã€‚
*   æœ€çµ‚çš„ã«ã¯ç”»åƒã‚’ä¸ãˆã‚‹ã¨ã€1~9ã®ã‚ˆã†ã«è­˜åˆ¥ã—ãŸã„ã®ã§ã€å‡ºåŠ›ã‚’ï¼‘æ¬¡å…ƒã«ã—ãŸã„ã€‚ãã®ãŸã‚ã«å…¨çµåˆå±¤ãŒã‚ã‚‹ã€‚
*   å…¨çµåˆå±¤ã®å‰ã«Flattenå±¤ã§å¹³å¦åŒ–ã™ã‚‹

```python
from keras.callbacks import EarlyStopping

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy"])

# å­¦ç¿’
model.fit(X_train, Y_train, epochs=1, batch_size=100, validation_data=(X_test, Y_test), callbacks=[EarlyStopping()])

# ãƒ¢ãƒ‡ãƒ«è©•ä¾¡
score = model.evaluate(X_test, Y_test, verbose=0)
print("loss:", score[0])
print("accuracy:", score[1])
```

## step3 æ¨è«–ã—ã¦ã¿ã‚‹

```python
from keras.models import load_model
import numpy as np
from matplotlib import pyplot as plt

# å­¦ç¿’æ¸ˆã¿ãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜ã—ã¦ãŠã
model.save('MNIST_NN.h5')
MNIST_NN = load_model('MNIST_NN.h5')

# å­¦ç¿’æ¸ˆã¿ãƒ¢ãƒ‡ãƒ«ã«ã€mnistã®ãƒãƒªãƒ‡ãƒ¼ã‚·ãƒ§ãƒ³ç”»åƒã‚’æ¸¡ã—ã¦æ¨è«–ã—ã¦ã¿ã‚‹
predicted = MNIST_NN.predict(x_test[0].reshape(1, 28, 28, 1))
print("åˆ¤å®šçµæœ=", np.argmax(predicted)) # æœ€ã‚‚ç¢ºç‡ã®é«˜ã„æ•°å­—ã‚’å‡ºåŠ›ã™ã‚‹
plt.imshow(x_test[0], cmap=plt.cm.gray_r)
plt.show
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from IPython.display import Image
import cv2
import numpy as np
from PIL import Image as PILImage
import io

# ã‚«ãƒ¡ãƒ©ã§å†™çœŸã‚’æ’®ã‚‹ãŸã‚ã®JavaScript
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
      async function takePhoto(quality) {
        const div = document.createElement('div');
        const capture = document.createElement('button');
        capture.textContent = 'ğŸ“· æ’®å½±';
        div.appendChild(capture);

        const video = document.createElement('video');
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});

        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();

        // è§£åƒåº¦
        google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

        await new Promise((resolve) => capture.onclick = resolve);

        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;
        canvas.getContext('2d').drawImage(video, 0, 0);
        stream.getTracks().forEach(track => track.stop());
        div.remove();

        return canvas.toDataURL('image/jpeg', quality);
      }
      takePhoto({quality: %f});
    ''' % quality)
    display(js)
    data = eval_js("takePhoto()")
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename
```

```python
from base64 import b64decode
from tensorflow.keras.models import load_model

# ç”»åƒã‚’æ’®å½±
photo_path = take_photo()

# ãƒ¢ãƒ‡ãƒ«ã‚’èª­ã¿è¾¼ã¿ï¼ˆå­¦ç¿’æ¸ˆã¿MNIST CNNãƒ¢ãƒ‡ãƒ«ï¼‰
model = load_model("MNIST_NN.h5")

# ç”»åƒã‚’èª­ã¿è¾¼ã‚“ã§å‰å‡¦ç†
img = PILImage.open(photo_path).convert('L')  # ã‚°ãƒ¬ãƒ¼ã‚¹ã‚±ãƒ¼ãƒ«
img = img.resize((28, 28))  # ãƒªã‚µã‚¤ã‚º
img_array = np.array(img)
img_array = 255 - img_array  # ç™½é»’åè»¢
img_array = img_array.astype('float32') / 255.0
img_array = img_array.reshape(1, 28, 28, 1)

# æ¨è«–
pred = model.predict(img_array)
predicted_digit = np.argmax(pred)

# çµæœè¡¨ç¤º
print("ğŸ“Œ äºˆæ¸¬ã•ã‚ŒãŸæ•°å­—:", predicted_digit)
```
