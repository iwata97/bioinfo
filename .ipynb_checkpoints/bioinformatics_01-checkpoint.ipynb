```python
import pandas as pd
import numpy as np
```

TPMをファイルから抽出する。

```python
df_tpm = pd.read_excel('https://github.com/iwata97/bioinfo/raw/refs/heads/main/TPM_data.xlsx')
```

中身を確認してみる

```python
df_tpm
```

全てを表示すると場所を取るので、上位5行だけ表示する

```python
df_tpm.head()
```

geneを削除

```python
tpm = df_tpm.drop('gene', axis=1)
tpm.head()
```

```python
tpm.iloc[:,0]
```

TPMデータでヒストグラムを描いてみる

```python
import matplotlib.pyplot as plt

x = tpm.iloc[:, 0] #batch_1のTPMデータでヒストグラムを作成する。
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(x, bins='scott')  # 階級の範囲の計算方法を指定する　#スコットの選択を検索
ax.set_xlabel('TPM')
ax.set_ylabel('frequency')
fig.show()
```

ばらつきの大きいデータであることがわかる

そこで、TPM値を対数化する。TPM値が0のときは対数化でエラーがでるため、あらかじめ1.0をたしておく。

```python
log_tpm = np.log10(tpm + 1.0)
log_tpm.head()
```

```python
x = log_tpm.iloc[:, 0]
fig = plt.figure()
ax = fig.add_subplot()
ax.hist(x, bins='scott')  # 階級の範囲の計算方法を指定する
ax.set_xlabel('log10(TPM)')
ax.set_ylabel('frequency')
fig.show()
```

同じAxesクラスのインスタンスに対してhist()メソッドを複数実行することで、同じ作図領域に複数のヒストグラムを描くことができる。

この際、ヒストグラムが重なるので、hist()メソッドのalpha引数に透明度を指定する。

対数TPMデータで、batch_1〜3とchemostat_1〜3の6データを重ねて描く。

図からわかるように０に近い値が多くある分布は対数変換することによって正規化する。

複数の変数間（batch_1とbatch_2、batch_1とbatch_3、etc.）の総当たりでヒストグラム付きの散布図を描くのは、Seabornというライブラリを使うと描ける。

```python
import seaborn as sns
sns.pairplot(log_tpm, diag_kind='kde') #diag_kind="kde"とすることで対角線に表示されるグラフの種類をカーネル密度推定に変更することができる。
plt.show()
```

今回は各遺伝子におけるchemostatの平均値とbatchの平均値を散布図にした。

```python
batch_log_tpm_mean = np.log2(tpm.iloc[:, :3].mean(axis=1).values + 1)
chemostat_log_mean = np.log2(tpm.iloc[:, 3:].mean(axis=1).values + 1)

log_tpm_lib_mean = pd.DataFrame({'batch': batch_log_tpm_mean,
                                 'chemostat': chemostat_log_mean})

sns.jointplot(data=log_tpm_lib_mean, x = 'batch', y = 'chemostat')
plt.show()
```
