### 糖尿病のデータを用いて解析を再現する
bioinformatics_2-1を参考にして、解析をしてみよう

scikit-learnにサンプルデータとしてあるDiabetes Dataset。糖尿病患者 442 人のデータとなっており、目的変数として1年後の「進行状況」と，説明変数として年齢、性別やBMIなどの下記10項目のデータ構成となっている。
機械学習はランダムフォレストで、回帰モデルを使う。
ライブラリは

from sklearn.ensemble import RandomForestRegressor

をつかう。

関連ライブラリの読み込み

```python
import pandas as pd
import numpy as np
```

```python
# データの読み込み
df_diabetes = pd.read_excel('https://github.com/iwata97/bioinfo/raw/refs/heads/main/diabetes_data.xlsx')
```

データの中身を見てみる

```python
df_diabetes
```

```python
# 説明変数と目的変数をそれぞれ読み込み
X = df_diabetes.drop('target', axis=1).values
y = df_diabetes['target'].values.ravel()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
# 特徴量の正規化
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# モデル評価結果保存用リスト
results = []
```

```python
from sklearn.ensemble import RandomForestRegressor

rand_forest = RandomForestRegressor(random_state=1)
rand_forest.fit(X_train, y_train)

# クロスバリデーション (5-fold)
scores = cross_val_score(rand_forest, X_train, y_train, cv=5)

# 平均と標準誤差（標準偏差 / √n）
rand_forest_acc = scores.mean()
rand_forest_std_err = scores.std(ddof=1) / np.sqrt(len(scores))

# 精度の保存
results.append({
    "Model": "Random Forest",
    "Accuracy": rand_forest_acc,
    "StdErr": rand_forest_std_err
})
```

```python
print(results)
```
